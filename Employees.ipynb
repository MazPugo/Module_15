```python
#2) Jakie ÅºrÃ³dÅ‚a pozyskania pracownika (Recruitment Source) sÄ… najlepsze, jeÅ›li zaleÅ¼y nam na jak najdÅ‚uÅ¼szym staÅ¼u pracownikÃ³w?
#(What are the best sources of recruiting an employee ( Recruitment Source) if we want the longest tenure of employees)
```

```python
df[['DateofTermination','DateofHire','RecruitmentSource']].dtypes
```

```python
df3=df[['DateofTermination','DateofHire','RecruitmentSource']]
df3.dropna(subset=['DateofTermination'],inplace = True)
```

```python
df3['year_T'] = pd.DatetimeIndex(df3['DateofTermination']).year
df3['year_H'] = pd.DatetimeIndex(df3['DateofHire']).year
df3
```

```python
df3.insert(3, "seniority", np.array(0), True)
df3
```

```python
df3['seniority'] = df3['year_T'].astype(int) - df3['year_H'].astype(int)
df3
```

```python
df10=df3.groupby(['RecruitmentSource']).sum()
df10
```

```python
numpy_array_seniority = df10['seniority'].to_numpy()
numpy_array_RecruitmentSource= df10.index.to_numpy()
numpy_array_seniority
```

```python
numpy_array_RecruitmentSource
```

```python
long_seniority = {numpy_array_seniority[i]: numpy_array_RecruitmentSource[i] for i in range(len(numpy_array_seniority))}
```

```python
long_seniority
```

```python
f, ax = plt.subplots(figsize=(20,8)) 

plt.bar(list(long_seniority.values()),list(long_seniority.keys()), align='center')
plt.xticks(range(len(long_seniority)), list(long_seniority.values()),rotation=80)
plt.title("the best sources of recruitment")
plt.xlabel("RecruitmentSource")
plt.ylabel("seniority")
plt.show()
```

```python
#Najlepszym zrodlem pozyskiwania pracownikow jest "Diversity Job Fair" a najgorszym "On-line Web application"
#(the best source to get employees is "Diversity Job Fair", the worst source: "On-line Web application" )
```

```python
# 4)Jak wyglÄ…da struktura wieku aktualnie zatrudnionych pracownikÃ³w?
# What is the age structure of currently employed employees?
```

```python
df["DOB"]
```

```python
from datetime import date,datetime
from dateutil.relativedelta import relativedelta
```

```python
w=date.today()
w=w.strftime("%Y-%m-%d")
def days_between(d1, d2):
    d1 = datetime.strptime(d1, "%Y-%m-%d")
    d2 = datetime.strptime(d2, "%Y-%m-%d")
    return abs(relativedelta(d2, d1).years)
```

```python
d=pd.to_datetime(df["DOB"]).apply(lambda x:x.strftime("%Y-%m-%d"))
df["Age"]=d.apply(lambda x:days_between(x, w))
df["Age"]
```

```python
plt.figure(figsize=(8,6))
sns.set_theme(style="darkgrid")
plt.grid(True)
ax=sns.kdeplot(df["Age"],shade=True,color="red")
ax.grid(which='major', axis='x', zorder=0, color='#DFFF00', linewidth=0.4)
ax.grid(which='major', axis='y', zorder=0, color='#DFFF00', linewidth=0.4)
plt.rc('font', size=10,weight="bold")
plt.xticks(rotation=20)
plt.tight_layout()
```

```python
#Zasadnicza wiekszosc pracownikow ma powyzej 30 lat, najmniej jest pracownikow powyzej 60 lat
#(the majority of employees is older than 30 years old,there are less employees over 60 years old)
```
